<p style="background-color:#E51A59; font-family:newtimeroman; color:#FDFEFE; font-size:175%; text-align:left; border-radius:0px 0px;">👉 Some Reminders on Exploratory Data Analysis (EDA) 👈</p>

<p dir="ltr">Exploratory data analysis (EDA) is an especially important activity in the routine of a data analyst or scientist. It enables an in depth understanding of the dataset, define or discard hypotheses and create predictive models on a solid basis. It uses data manipulation techniques and several statistical tools to describe and understand the relationship between variables and how these can impact business. By means of EDA, we can obtain meaningful insights that can impact analysis under the following questions (If a checklist is good enough for pilots to use every flight, it’s good enough for data scientists to use with every dataset).</p>

<ol type="I">
    <li>What question are you trying to solve (or prove wrong)?</li>
    <li>What kind of data do you have?</li>
    <li>What’s missing from the data?</li>
    <li>Where are the outliers?</li>
    <li>How can you add, change or remove features (Feature Engineering) to get more out of your data?</li>
</ol>

<p dir="ltr"><b>Exploratory data analysis (EDA)</b> is often an <b>iterative brainstorming process</b> where you pose a question, review the data, and develop further questions to investigate before beginning model development work.
```python
df = df.rename(columns={'Body type': 'body_type',
                        'Model code' : 'model_code',
                        'Country version': 'country_version',
                        'Offer number': 'offer_number',
                        'First registration' : 'first_registration',
                        'Fuel type': 'fuel_type',
                        'Engine size': 'engine_size',
                        'CO₂-emissions': 'co2_emissions',
                        'Manufacturer colour': 'manu_color',
                        'Fuel consumption': 'fuel_consumption',
                        '\nComfort & Convenience\n': 'comfort_convenience',
                        '\nEntertainment & Media\n':'ent_media',
                        '\nSafety & Security\n': 'safety_security',
                        '\nExtras\n': 'extras',
                        'Empty weight': 'empty_weight',
                        'General inspection': 'gen_inspection',
                        'Last service' : 'last_service',
                        'Full service history': 'service_history',
                        'Non-smoker vehicle': 'non_smoker',
                        'Emission class': 'emission_class',
                        'Emissions sticker':'emission_sticker',
                        'Upholstery colour': 'upholstery_color',
                        'Production date': 'pro_date',
                        'Previous owner': 'previous_owner',
                        'Other fuel types':'other_fuel',
                        'Power consumption': 'power_cons',
                        'Energy efficiency class': 'energy_class',
                        'CO₂-efficiency': 'co2_efficiency',
                        'Fuel consumption (WLTP)': 'fuel_wltp',
                        'CO₂-emissions (WLTP)': 'co2_wltp',
                        'Available from': 'available',
                        'Taxi or rental car': 'taxi_rental',
                        'Last timing belt change': 'last_timing_belt',
                        'Electric Range (WLTP)': 'elec_range_wltp',
                        'Power consumption (WLTP)': 'power_cons_wltp',
                        'Battery Ownership': 'battery_owner',
                        'Type' : 'type'                      
                        
                       })
```

```python
df.columns = df.columns.str.lower()
```

```python
# Function to clean a column with list-like string data
def clean_list_column(column):
    return column.astype(str).str.replace(r'^[,\s]+|[,\s]+$', '', regex=True)

# Clean specific columns
columns_to_clean = ['model', 'body_type', 'type', 'doors', 'country_version', 'offer_number', 'warranty', 'gearbox','gears']
for col in columns_to_clean:
    if col in df.columns:
        df[col] = clean_list_column(df[col])

# Clean the 'make' column to remove newline characters
if 'make' in df.columns:
    df['make'] = df['make'].astype(str).str.replace(r'\n', '', regex=True).str.strip(", ")

# Convert price to numeric (remove currency symbol and commas)
#if 'price' in df.columns:
    #df['price'] = df['price'].astype(str).str.replace(r'[^0-9,.-]', '', regex=True).str.replace(',', '').replace('', np.nan).astype(float)

# Convert mileage to numeric (remove " km" and commas)
if 'mileage' in df.columns:
    df['mileage'] = df['mileage'].astype(str).str.replace(r'[^0-9,.-]', '', regex=True).str.replace(',', '').replace('', np.nan).astype(float)

```

```python
#clean model column
def strip_chars(column):
    return column.str[5:-5]

# Apply the function to the specified columns
columns_to_strip = ['model']
for col in columns_to_strip:
    if col in df.columns:
        df[col] = strip_chars(df[col])
```

```python
#clean price column
df['price'] = df['price'].str.replace(r'[^\d,]', '', regex=True).str.replace(',', '').astype(float)
```

```python
#clean body_type
def strip_chars(column):
    return column.str[5:-6]

# Apply the function to the specified columns
columns_to_strip = ['body_type']
for col in columns_to_strip:
    if col in df.columns:
        df[col] = strip_chars(df[col])
```

```python
#clean 'type','doors','country_version','offer_number','warranty' columns
def strip_chars(column):
    return column.str[5:-6]

# Apply the function to the specified columns
columns_to_strip = ['type','doors','country_version','offer_number','warranty']
for col in columns_to_strip:
    if col in df.columns:
        df[col] = strip_chars(df[col])
```

```python
#clean gearbox column
def strip_chars(column):
    return column.str[4:-4]

# Apply the function to the specified columns
columns_to_strip = ['gearbox']
for col in columns_to_strip:
    if col in df.columns:
        df[col] = strip_chars(df[col])
```

```python
#clean these columns
column_list = ['seats','engine_size','gears','co2_emissions','cylinders','empty_weight']
```

```python
for col in column_list:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace(r'[^0-9,.-]', '', regex=True).str.replace(',', '').replace('', np.nan)
        df[col] = df[col].astype(float)
```

```python
def convert_warranty(value):
    if pd.isna(value) or value.strip() == '':
        return np.nan
    value = value.strip()
    if value.lower() == 'yes':
        return 'yes'
    elif 'months' in value:
        return float(value.replace('months', '').strip())
    else:
        return np.nan

df['warranty'] = df['warranty'].apply(convert_warranty)
```

```python
#clean drivetrain column
if 'drivetrain' in df.columns:
    df['drivetrain'] = df['drivetrain'].astype(str).str.extract(r'(4WD|Rear|Front)')
```

```python
#clean emission_sticker column
# Define a function to extract text within parentheses
def extract_text(s):
    if pd.notna(s):  # Ensure the element is not NaN
        match = re.search(r'\(([^)]+)\)', s)  # Find text within parentheses
        if match:
            return match.group(1)
    return np.nan  # Return NaN if no match or if input is NaN

# Apply the function to the 'emission_sticker' column
df['emission_sticker'] = df['emission_sticker'].apply(extract_text)
```

```python
# Create a boolean mask for rows where one column is not null and the other is null
mask = (df['emission_sticker'].notnull() & df['emission_class'].isnull()) | (df['emission_sticker'].isnull() & df['emission_class'].notnull())

# Apply the mask to get the filtered DataFrame
result = df[mask]
```

```python
df['pro_date'] = pd.to_datetime(df['pro_date'].dropna().astype(int), format='%Y', errors='coerce')
```

```python
# divide power column into two column
df['power'] = df['power'].astype(str)  # Convert to string type
df['power'] = df['power'].str.replace(r'\[|\]', '', regex=True)  # Remove square brackets
df[['power_kW', 'power_hp']] = df['power'].str.extract(r'(\d+) kW \((\d+) hp\)').astype(float)  # Extract kW and hp into separate columns
```

```python
# Drop the 'power' column
df.drop(columns=['power'], inplace=True)
```

```python
#clean model code
if 'model_code' in df.columns:
    df['model_code'] = df['model_code'].astype(str).str.replace(r'[^0-9A-Z,.-/]', '', regex=True).str.replace(',', '').replace('', np.nan)
```

```python
#extract fuel consumption as average-comb
df['fuel_consumption'] = df['fuel_consumption'].astype(str)  # Convert to string type
df['fuel_consumption'] = df['fuel_consumption'].str.replace(r'[\[\]]', '', regex=True)  # Remove square brackets

# Extract the combined fuel consumption values in liters per 100 km
df['fuel_consumption_comb'] = df['fuel_consumption'].str.extract(r'(\d+\.\d+) l/100 km \(comb\.\)').astype(float)
df['fuel_consumption_city'] = df['fuel_consumption'].str.extract(r'(\d+\.\d+) l/100 km \(city\)').astype(float)
df['fuel_consumption_country'] = df['fuel_consumption'].str.extract(r'(\d+\.\d+) l/100 km \(country\)').astype(float)
```

```python
# Drop the 'fuel_consumption' column
df.drop(columns=['fuel_consumption'], inplace=True)
```

```python
# Split location and assign country and city to DataFrame

df['location'] = df['location'].astype(str)

df['location_country'] = ""
df['location_city'] = ""

for index, location in enumerate(df['location']):
    parts = location.split(", ")
    if len(parts) >= 2:  # Ensure there are at least two parts (city and country)
        df.at[index, 'location_country'] = parts[-1]  # Last part is the country
        df.at[index, 'location_city'] = parts[-2]     # Second-to-last part is the city
```

```python
df['location_city'] = df['location_city'].apply(lambda x: ' '.join(x.split()[1:]) if len(x.split()) > 1 else '')
```

```python
def list_to_string(column):
    return column.apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

columns_to_convert = ['ent_media', 'desc', 'comfort_convenience', 'safety_security', 'extras']  # Dönüştürmek istediğiniz sütunların adları

for col in columns_to_convert:
    if col in df.columns:
        df[f'{col}_str'] = list_to_string(df[col])
```

```python
df.loc[df['fuel_type'].str.contains('Super', case=False, na=False), 'fuel_type'] = 'Benzine'
df.loc[df['fuel_type'].str.contains('Gasoline', case=False, na=False), 'fuel_type'] = 'Benzine'

# İçinde 'gas' geçenleri 'Gasoline' olarak güncelle
df.loc[df['fuel_type'].str.contains('Gas', case=False, na=False), 'fuel_type'] = 'LPG'

# İçinde 'diesel' geçenleri 'Diesel' olarak güncelle
df.loc[df['fuel_type'].str.contains('Diesel', case=False, na=False), 'fuel_type'] = 'Diesel'
df.loc[df['fuel_type'].str.contains('Benzine', case=False, na=False), 'fuel_type'] = 'Benzine'
df.loc[df['fuel_type'].str.contains('Electric', case=False, na=False), 'fuel_type'] = 'Electric'
df.loc[df['fuel_type'].str.contains('LPG', case=False, na=False), 'fuel_type'] = 'LPG'
df.loc[df['fuel_type'].str.contains('CNG', case=False, na=False), 'fuel_type'] = 'LPG'
df.loc[df['fuel_type'].str.contains('Hydrogen', case=False, na=False), 'fuel_type'] = 'Electric'
df.loc[df['fuel_type'].str.contains('Ethanol', case=False, na=False), 'fuel_type'] = 'Benzine'
df.loc[df['fuel_type'].str.contains('Particle filter', case=False, na=False), 'fuel_type'] = 'Diesel'
df.loc[df['fuel_type'].str.contains('Others', case=False, na=False), 'fuel_type'] = 'Benzine'
```

```python
# 'previous_owner' sütunundaki liste elemanlarını ayırma
def split_previous_owner(entry):
    if isinstance(entry, list) and isinstance(entry[0], list) and len(entry[0]) > 1:
        km = entry[0][0]
        date = entry[0][1]
        owner_count = entry[1] if len(entry) > 1 else None
    elif isinstance(entry, list) and len(entry) > 1:
        km = entry[0]
        date = None
        owner_count = entry[1]
    else:
        km = None
        date = None
        owner_count = entry if isinstance(entry, (int, float)) else None
    return km, date, owner_count
df[['prev_km', 'prev_date', 'owner_count']] = df['previous_owner'].apply(split_previous_owner).apply(pd.Series)
# 'prev_km' sütunundaki değerleri sayı formatına dönüştürme
df['prev_km'] = df['prev_km'].str.replace(',', '').str.replace(' km', '').astype(float, errors='ignore')
```
